### Comparing several machine learning algorithms using lazy predict library

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import lazypredict
from lazypredict.Supervised import LazyRegressor
```

## **Load the data set**

```python
df = pd.read_csv("Bioactivity_data_3class_pIC50_pubchem_fp.csv")
```

## **Splitting into X and Y variables**

```python
X = df.drop("pIC50", axis=1)
Y = df.pIC50
```

## **Data pre-processing**

```python
# Examine X dimension ( compounds, descriptors)
X.shape
```

```python
# Remove low variance features
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))  # Formula for binary(0 and 1) is p*(1-p); p: How many 1's and (1-p): how many 0's, p * 1-p tells how mixed they are
X = selection.fit_transform(X) # fit calculates the variance using p*(1-p) and transform filters the columns based on given variance threshold.
X.shape
```

```python
# Perform data splitting using 80/20 ratio
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

## **Compare ML algorithms**

```python
# Define and builds the lazyclassifier
# Model building uses default parameters for all of the algorithms used.
clf = LazyRegressor(verbose=0, ignore_warnings = True, custom_metric = None)
train, test = clf.fit(X_train, X_test, Y_train, Y_test)
```
