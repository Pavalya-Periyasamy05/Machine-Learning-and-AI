# **Linear Regression and Random Forest**

# **Load data**

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df
```

# Data Preparation

## Data split or Data Separation as X and Y

```python
y = df["logS"]
y
```

```python
x = df.drop("logS", axis=1)
```

## Data splitting into Train and Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# **Model Building**

## **Linear Regression**

### Training the model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
x_train.shape # Linearregression function looks at the shape of the variable if it has got multiple variable or one single variable (xsamples, xfeatures)
```

### Applying the model to make prediction

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### Evaluate model performance

```python
y_train
```

```python
y_lr_train_pred # predicted value
```

```python

from sklearn.metrics import mean_squared_error, r2_score

# Training set
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred) # squared correlation coefficient

# Testing set
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

# MSE- First finds the difference between true and pred value for each rows and then converts the diff values into square( to avoid cancelling out due to + and - directions) and then
#finally does the average mean, less value shows the error is less.
# R2 - Tells you if the model just focused on Y variable average or is it taken the input of X variable to predict the y variable to find the pattern.
# MSE and R2 standard threshold in word document(documents/cambridge/Linear Regression and Random Forest ML.docx)
```

```python
print("LR MSE (Train) :" ,lr_train_mse)
print("LR R2 (Train) :", lr_train_r2)
print("LR MSE (Test) :" ,lr_test_mse)
print("LR R2 (Test) :", lr_test_r2)
```

```python
lr_results = pd.DataFrame(["Linear regression", lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns=["Method", "Training MSE", "Training R2", "Test MSE", "Test R2"]
```

```python
lr_results
```

## **Random Forest**

### Training the model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100) #
rf.fit(x_train, y_train)
```

### Applying the model to make prediction

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

# Training set
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred) # squared correlation coefficient

# Testing set
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(["Random Forest",rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2] ).transpose() # Rows becomes columns and columns becomes rows
rf_results.columns =["Method", "Training MSE", "Training R2", "Test MSE", "Test R2"]
rf_results
```

## **Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0) # concat: To comnbine two dataframes vertically or horizontally (rows, axis=0)
df_models
```

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
# or
df_models.reset_index(drop=True)
df_models
```

## **Data Visualization of prediction results**

```python
#If the dispersion is low we expect that performance will be good
# Taking the predicted value and actual value using scatter plot
import matplotlib.pyplot as plt
import numpy as np # To get the trend line

plt.figure(figsize=(5,5))
plt.scatter(x = y_train, y= y_lr_train_pred, c= "#7CAE00", alpha =0.3 ) # alpha: makes overlapping look darker and not overlapping look lighter, c= "#7CAEOO" for color
# green points only represents the predicted data points, it is always shown on y axis
# Trend line is what represents the true data
# For plotting predicted data points or drawing trend line we need x and Y axis to find slope and intercept
#X-axis: experimental/actual values
#Y-axis: predicted values
#The intercept is the point on the line where X = 0 corresponding to Y.
#Important: X = 0 is usually not an actual data point — it’s an imaginary/reference point.
#The intercept helps you understand the direction of the line:
#Line going upward after the intercept → predictions increase with X
#Line going downward after the intercept → predictions decrease with X
#In other words, the intercept is mainly used to determine the slope and overall trend, not to represent real data.

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)
plt.plot(y_train, p(y_train),"#F8766D")

plt.xlabel("Experimental logs")
plt.ylabel("Predicted logs")
plt.plot()
```
